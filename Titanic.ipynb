#### But what if we and wanted to see just the people from Cherbourg?
Well then, we can slice the datafram using a special type of object notation

```python
titanic_df[titanic_df.Embarked =="C"]
```

#### What about if we wanted to further segment this to just include men from Cherborg who survived?

```python
titanic_df[(titanic_df.Embarked =="C") & (titanic_df.Sex =="male")]
```
