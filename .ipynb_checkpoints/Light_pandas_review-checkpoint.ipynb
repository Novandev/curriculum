### Part of the data science process is to manipulate the data to ask different questions in order to do more investigating: lets sat that we wanted to know survival rates....but based on which deck they were on? A little bit of investigating the data tells us that the first character in cabin for any row denotes the deck level that a given passenger was on. We'll start by just indexing the Cabin column from the dataframe so its isolated

```python
cabin_series = titanic_df['Cabin']
# to recap titanic_df.Cabin is also acceptable

# just to make sure that the types are correct, lets check the types
type(cabin_series)
```

#### The next part involves us iterating over the cabin_series object that we created. Just like in python list, the Series can be used in a for loop to access its values
