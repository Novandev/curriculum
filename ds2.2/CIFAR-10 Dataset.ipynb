#### Scalar
Scalars are data that are fully described by numerical value alone.
Think of an int type (or Number in Javascript). It is simple a value that describes a single quantity

<b>Programatically scalars would look like this:</b>

```python
#python
var_count = 0
```

```javascript
#javascript
let varCount = 0;
```

```
#Swift
let varCount: Int = 10;
```

*USE IN DATA SCIENCE:*
In your dataset, this would be a value of one column of a singlar

One thing to note is that there is *no other form of measurement for this datatype* meaning that the datatype cant be indexed....Speaking of indexable datatypes...


#### Vectors
Vectors are data that are fully described by a list of numerical values.

If you just skowled at this page becuase the word vector reminded you of Vector Calculus and all those wild symbols, you're not alone. 
Think of vectors as a container for scalars. 

<b>Programatically vectors would look like this:</b>

```python
#python
list_of_counts = [1,2,3,5,6,7]
```

```javascript
#javascript
let arrayOfCounts = [1,2,3,5,6,7];
```

#### Tensors and dimensions
Tensors have the same relationship to vectors in terms of magnitude as vectors have to scalars. That is, it has a higher number of dimensions that values exist in.

Thinking programatically instead of mathematically:
    - a scalar cant be indexed, since itself *IS* a measurement
        - think of how when you call a variable of this type, it just returns its value
    - A vector can be indexed by position but only with one index
